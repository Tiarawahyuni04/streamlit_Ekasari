```python

# Ambil 10 nilai TOEFL Score yang paling sering muncul
top_scores = data['TOEFL Score'].value_counts().nlargest(10).index

# Filter data hanya untuk TOEFL Score yang ada di top_scores
filtered_data = data[data['TOEFL Score'].isin(top_scores)]

# Membuat plot countplot dengan data yang sudah difilter
f = plt.figure(figsize=(10,5))
sns.countplot(x='TOEFL Score', data=filtered_data, palette='Spectral')
plt.xticks(rotation=90)
plt.show()
```

```python
#data preparation
x = data.drop('TOEFL Score', axis=1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['TOEFL Score'] = le.fit_transform(data['TOEFL Score'])
```

```python
data.head()
```

```python
y = data['TOEFL Score']
```

```python
y.head()
```

```python
#train model untuk modeling
model = []
accuracy = []
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier()
```

```python
# Lanjutkan dengan fitting model
DT.fit(x_train, y_train)
```

```python
predict = DT.predict(x_test)
```

```python
DT_accuracy = DT.score(x_test, y_test)
```

```python
DT_accuracy
```

```python
accuracy.append(DT_accuracy)
model.append('Decision Tree')
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
LG = LogisticRegression()
```

```python
LG.fit(x_train, y_train)
```

```python
LG_accuracy = LG.score(x_test, y_test)
```

```python
LG_accuracy 
```

```python
accuracy.append(LG_accuracy )
model.append('Logistic Regression')
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier()
```

```python
RF.fit(x_train, y_train)
```

```python
RF_accuracy = LG.score(x_test, y_test)
```

```python
RF_accuracy
```
